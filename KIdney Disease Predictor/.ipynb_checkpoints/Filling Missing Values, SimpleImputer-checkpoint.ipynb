# Filling Missing Values



# Load Dataset

```python
import pandas as pd 
df = pd.read_csv("binary_classification_sample.csv")
df
```

# Filling Missing Values

```python
df.isnull().sum()
```

```python
# fill numeric columns with mean
from sklearn.impute import SimpleImputer

impter1 = SimpleImputer(strategy='mean')
df[['Salary','LocationScore']] = impter1.fit_transform(df[['Salary','LocationScore']])
```

```python
# fill categorical column with most frequent category
impter2 = SimpleImputer(strategy='most_frequent')
df[['Education']] = impter2.fit_transform(df[['Education']])
```

```python
df.isnull().sum()
```
