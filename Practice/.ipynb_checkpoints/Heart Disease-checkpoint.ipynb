```python
X = df.drop('target', axis=1)
```

```python
Y = df.target
```

```python
clf = RandomForestClassifier()
rgf = RandomForestRegressor()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
rclf = clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
rgf.fit(x_train,y_train)
```

```python
rgf.score(x_test,y_test)
```

here we have a classification dataset but regression is also working becaurse of numeric encoding

```python
cross_val_score(rclf,X,Y,cv=5)
```

```python
y_proba = rclf.predict_proba(x_test)
```

```python
y_prob_positive = y_proba[:,1]
```

```python
y_prob_positive
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
fpr, tpr, thresholds = roc_curve(y_test,y_prob_positive)
```

```python
fpr
```

```python

def plot_roc(fpr,tpr):
    # Plot Roc Curve
    plt.plot(fpr, tpr, color="red", label="AUC")
    plt.legend()
    plt.show()

plot_roc(fpr,tpr)
    
```
